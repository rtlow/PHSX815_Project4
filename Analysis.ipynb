```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sep, dsep = np.loadtxt('separation_distances.txt')
```

```python
plt.figure(figsize=[12,7])
bins = int(len(sep) / 10)
avg = np.mean(sep)
sqrtmeansqerr = np.sqrt(np.mean(dsep**2))
plt.hist(sep, bins=bins)
plt.xlabel('Separation Distance')
plt.ylabel('Frequency')
plt.title('Histogram of Separation Distances')
plt.axvline(avg, color='k')
plt.plot([],[], label='Average separation distance: {:.2f}'.format(avg), color='k')
plt.plot([],[], label='Mean Error: {:f}'.format(sqrtmeansqerr), alpha=0)
plt.legend()
plt.savefig('sepdist.png')
plt.show()
```
